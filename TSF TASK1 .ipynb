```python
# IMPORTING THE REQUIRED LIBRARIES.
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error
```

```python
# READING THE DATA.
data = pd.read_csv("http://bit.ly/w-data")
data.head(5)
```

```python
# CHECKING IF THERE IS ANY NULL VALUE IN THE DATASET.
data.isnull == True
```

```python
#As there is no null value in our dataset,so,we can now visualize our data.
```

```python
#PLOTTING THE DISTRIBUTION OF SCORES.
sns.set_style('darkgrid')
sns.scatterplot(y= data['Scores'], x= data['Hours'])
plt.title('Marks Vs Study Hours',size=20)
plt.ylabel('Marks Percentage', size=12)
plt.xlabel('Hours Studied', size=12)
plt.show()
```

```python
#From the above scatter plot,we can see that there is a co-relation between "Marks" and "Hours Studied".
#So,now we can plot a regression line to confirm the co-relation.
```

```python
#Plotting the Regression.
sns.regplot(x= data['Hours'], y= data['Scores'])
plt.title('Regression Plot',size=20)
plt.ylabel('Marks Percentage', size=12)
plt.xlabel('Hours Studied', size=12)
plt.show()
print(data.corr())
```

```python
#From the above regression plot we can confirm that the two(2) variables that is "Marks" and "Hours Studied" are co-related.
```

```python
#Training Model
```

```python
#STEP-1 : Splitting the Data.
# Defining X and y from the Data
X = data.iloc[:, :-1].values  
y = data.iloc[:, 1].values

# Spliting the Data in two
train_X, val_X, train_y, val_y = train_test_split(X, y, random_state = 0)
```

```python
#STEP-2 : Fitting the Data into the model.
regression = LinearRegression()
regression.fit(train_X, train_y)
print("---------Model Trained---------")
```

```python
#Predicting the Percentage of Marks
pred_y = regression.predict(val_X)
prediction = pd.DataFrame({'Hours': [i[0] for i in val_X], 'Predicted Marks': [k for k in pred_y]})
prediction
```

```python
#Comparing the predicted marks with actual marks.
compare_scores = pd.DataFrame({'Actual Marks': val_y, 'Predicted Marks': pred_y})
compare_scores
```

```python
#Visusally comparing the predicted and actual marks.
plt.scatter(x=val_X, y=val_y, color='blue')
plt.plot(val_X, pred_y, color='Black')
plt.title('Actual vs Predicted', size=20)
plt.ylabel('Marks Percentage', size=12)
plt.xlabel('Hours Studied', size=12)
plt.show()
```

```python
# Calculating the accuracy of the model
print('Mean absolute error: ',mean_absolute_error(val_y,pred_y))
```

```python
#Small value of Mean absolute error states that the chances of error or wrong forecasting through the model are very less.
```

```python
#What will be the predicted score of a student if he/she studies for 9.25 hrs/ day?
```

```python
hours = [9.25]
answer = regression.predict([hours])
print("Score = {}".format(round(answer[0],3)))
```

```python
#So,according to the regression model if a student studies for 9.25 hours a day he/she is likely to score 93.893 marks.
```
